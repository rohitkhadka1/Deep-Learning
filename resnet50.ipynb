```python
import keras
from keras.applications.resnet50 import ResNet50
from keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt
```

```python
# Load image from URL
url = r'https://imgs.search.brave.com/7jlp1ziegFpG0KQhsvDCUxdTxCRowJ3Jpmvhw-mQK9U/rs:fit:500:0:0:0/g:ce/aHR0cHM6Ly9wcmV2/aWV3LnJlZGQuaXQv/bXktZmF2b3VyaXRl/LXRpZ2VyLWltYWdl/cy12MC1ldG0wZmVn/dm02ZGUxLmpwZz93/aWR0aD02NDAmY3Jv/cD1zbWFydCZhdXRv/PXdlYnAmcz0xMjhk/NGIwNmExY2E0ZDkw/YzZkYmNkZTA4NGI3/OGQ5MDIxNGQ1NDQ1'  # Replace with actual image URL
response = requests.get(url)
img = Image.open(BytesIO(response.content)).resize((224, 224))
```

```python
# Display the image using matplotlib
plt.imshow(img)
plt.axis('off')  # Hide axis ticks and labels
plt.show()
```

```python
model = ResNet50(weights = "imagenet")
x = keras.utils.img_to_array(img)
x = np.expand_dims(x, axis = 0)
x = preprocess_input(x)

preds = model.predict(x)
```

```python
print("Predicted: ", decode_predictions(preds, top = 1)[0])
```
